```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import requests
from census import Census
from pprint import pprint
```

```python
# Pull in all crime data from 2000 - 2020
crime_2000 = pd.read_csv("../Resources/Crime Information 2000.csv")
crime_2001 = pd.read_csv("../Resources/Crime Information 2001.csv")
crime_2002 = pd.read_csv("../Resources/Crime Information 2002.csv")
crime_2003 = pd.read_csv("../Resources/Crime Information 2003.csv")
crime_2004 = pd.read_csv("../Resources/Crime Information 2004.csv")
crime_2005 = pd.read_csv("../Resources/Crime Information 2005.csv")
crime_2006 = pd.read_csv("../Resources/Crime Information 2006.csv")
crime_2007 = pd.read_csv("../Resources/Crime Information 2007.csv")
crime_2008 = pd.read_csv("../Resources/Crime Information 2008.csv")
crime_2009 = pd.read_csv("../Resources/Crime Information 2009.csv")
crime_2010 = pd.read_csv("../Resources/Crime Information 2010.csv")
crime_2011 = pd.read_csv("../Resources/Crime Information 2011.csv")
crime_2012 = pd.read_csv("../Resources/Crime Information 2012.csv")
crime_2013 = pd.read_csv("../Resources/Crime Information 2013.csv")
crime_2014 = pd.read_csv("../Resources/Crime Information 2014.csv")
crime_2015 = pd.read_csv("../Resources/Crime Information 2015.csv")
crime_2016 = pd.read_csv("../Resources/Crime Information 2016.csv")
crime_2017 = pd.read_csv("../Resources/Crime Information 2017.csv")
crime_2018 = pd.read_csv("../Resources/Crime Information 2018.csv")
crime_2019 = pd.read_csv("../Resources/Crime Information 2019.csv")
crime_2020 = pd.read_csv("../Resources/Crime Information 2020.csv")
```

```python
def crime_data(neighborhood):
    years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 
             2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
    neighborhood_df = pd.DataFrame()
    neighborhood_df = crime_2000.loc[(crime_2000["NEIGHBORHOOD_NAME"] == neighborhood)]
    neighborhood_df = neighborhood_df.append(crime_2001.loc[(crime_2001["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2002.loc[(crime_2002["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2003.loc[(crime_2003["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2004.loc[(crime_2004["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2005.loc[(crime_2005["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2006.loc[(crime_2006["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2007.loc[(crime_2007["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2008.loc[(crime_2008["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2009.loc[(crime_2009["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2010.loc[(crime_2010["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2011.loc[(crime_2011["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2012.loc[(crime_2012["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2013.loc[(crime_2013["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2014.loc[(crime_2014["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2015.loc[(crime_2015["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2016.loc[(crime_2016["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2017.loc[(crime_2017["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2018.loc[(crime_2018["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2019.loc[(crime_2019["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2020.loc[(crime_2020["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df["YEAR"] = years
    neighborhood_df = neighborhood_df[["YEAR", "NEIGHBORHOOD_NAME", "HOMICIDE", "SEX_OFFENSE", "ROBBERY",
                                       "ASSAULT", "BURGLARY", "VICE", "THEFT", "VEH_THEFT", "OTHER"]]
    #neighborhood_df = neighborhood_df.set_index("YEAR")
    return neighborhood_df
```

```python
# Call function to create a crime data DataFrame for a location
crime_df = pd.DataFrame()
crime_df = crime_data("CHURCH HILL")
crime_df["TOTAL CRIME"]=crime_df["HOMICIDE"]+crime_df["SEX_OFFENSE"]+crime_df["ROBBERY"]+crime_df["ASSAULT"]+crime_df["BURGLARY"]+crime_df["VICE"]+crime_df["THEFT"]+crime_df["VEH_THEFT"]+crime_df["OTHER"]
crime_df
```

```python
housing_df = pd.read_csv("../Resources/Church Hill Housing.csv")
#housing_df = housing_df.set_index("Year")
housing_df
```

```python

crime_housing_df=pd.merge(crime_df,housing_df,how="outer",on="YEAR")
crime_housing_df
```

```python
#Cleaning database to totals for statistics
crime_housing_df=crime_housing_df.loc[:,["YEAR","NEIGHBORHOOD_NAME","TOTAL CRIME","AVERAGE HOUSE PRICE"]]
crime_housing_df
```

```python
#line plot comparison

#crime_housing_df.plot(x=crime_housing_df['YEAR'], y=["TOTAL CRIME", "AVERAGE HOUSE PRICE"], kind="bar")

crime_housing_df[["YEAR", "TOTAL CRIME"]].plot(x="YEAR", kind="bar")

plt.xlabel('YEAR')
plt.ylabel("TOTAL CRIME")
plt.title(f'TOTAL CRIME change 2000-2020')
plt.show()
```

```python
crime_housing_df[["YEAR", "AVERAGE HOUSE PRICE"]].plot(x="YEAR", kind="bar")

plt.xlabel('YEAR')
plt.ylabel("AVERAGE HOUSE PRICE")
plt.title(f'"AVERAGE HOUSE PRICE" change 2000-2020')
plt.show()
```
